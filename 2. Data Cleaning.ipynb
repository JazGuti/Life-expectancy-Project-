## 2. Missing Values 

```python
# Checking for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Calculate the percentage of missing values in each column
missing_percentage = df.isnull().mean() * 100
print(missing_percentage)

# display only columns with more than a certain threshold of missing values
high_missing = missing_percentage[missing_percentage > 50]  # columns with more than 50% missing
print(high_missing)
```

## 3. Address Missing Values 

```python
# Drop columns with more than 50% missing values
df.drop(columns=high_missing.index, inplace=True)
```

```python
#Checking
df.head()
```

```python
# Select only numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Replace missing values in numerical columns with the median for each country
df[numerical_columns] = df.groupby('Country Name')[numerical_columns].transform(lambda group: group.fillna(group.median()))

# Verify the missing values are filled
print(df.isnull().sum())
```

```python


# Impute numerical columns with the median
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
```

```python
print(df.isnull().sum())
```

## 3. Renaming Values 

```python
# Rename 'IncomeGroup' to 'Income Group'
df.rename(columns={'IncomeGroup': 'Income Group'}, inplace=True)
```

```python
# Rename 'Life Expectancy World Bank' to 'Life Expectancy'
df.rename(columns={'Life Expectancy World Bank': 'Life Expectancy'}, inplace=True)
```

```python
# Rename the columns 'Health Expenditure %' and 'Education Expenditure %'
df.rename(columns={'Health Expenditure %': 'Health Expenditure',
                   'Education Expenditure %': 'Education Expenditure'}, inplace=True)
```

```python
# Rename 'Prevelance of Undernourishment' to 'Prevalence of Undernourishment'
df.rename(columns={'Prevelance of Undernourishment': 'Prevalence of Undernourishment'}, inplace=True)
```

## 4. Create a Categorical Column 

```python
# Create a new column comparing the two disease categories
df['Disease Type'] = df.apply(
    lambda row: 'Communicable' if row['Communicable'] > row['NonCommunicable'] else 'NonCommunicable', axis=1
)
```


## 6. Detecting Outliers

```python
import math

# Create boxplots for all numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the number of rows and columns needed for the subplots
n_cols = 3
n_rows = math.ceil(len(numerical_columns) / n_cols)

plt.figure(figsize=(15, 5 * n_rows))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(df[column])
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

Because life expectancy can vary throughout different regions outliers are significant due to events that each region undergoes.
