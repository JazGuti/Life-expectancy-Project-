## 6. Categorical Plots 

#### Creating a categorical variable that splits "Life Expectancy" column in categories.

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['Life Expectancy'], bins = 10, kde = True)
```

```python
df.loc[df['Life Expectancy'] < 60, 'LE category'] = 'Low LE'
```

```python
df.loc[(df['Life Expectancy'] >=60) & (df['Life Expectancy'] < 75), 'LE category'] = 'Average LE'
```

```python
df.loc[df['Life Expectancy'] >= 75, 'LE category'] = 'High LE'
```

```python
df['LE category'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="Sanitation", y="Life Expectancy", hue="LE category", data=df)
```

```python
# Bin the 'Sanitation' column into categories
df['Sanitation Category'] = pd.cut(df['Sanitation'], bins=[0, 25, 50, 75, 100], labels=['Low', 'Medium', 'High', 'Very High'])

# Create the catplot with the binned Sanitation categories
sns.catplot(x="Sanitation Category", y="Life Expectancy", hue="LE category", data=df)

# Add title and labels
plt.title("Sanitation Category vs. Life Expectancy by Life Expectancy Category")
plt.xlabel("Sanitation Category")
plt.ylabel("Life Expectancy")

# Show the plot
plt.show()
```

6. Create a categorical plot and interpret the results.

Sanitation is a strong predictor of life Expectancy. Countries with higher levels of Sanitation do not have low Life Expectancy. Low and medium sanitation levels have lowe life expectancy category.
